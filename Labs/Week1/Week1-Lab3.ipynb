```python
#Lab 4: Billing Calculator with Default Arguments

def calculate_total(amount,tax_rate=0.05):
    tax_amount=amount*tax_rate
    amount_after_tax=amount+tax_amount
    return amount_after_tax

print(f"Total: {calculate_total(100):.2f}")
print(f"Total: {calculate_total(200, 0.12):.2f}")
```

```python
#Lab 5: Function Naming and Docstring — Discount Calculator

def apply_discount(price,discount_percentage):
    """Returns the discount price after applying the discount percentage(which is discount/100)"""
    Discount_price=price*(1-discount_percentage/100)
    return Discount_price

apply_discount(250,10)
```

```python
#Lab 6: Reusable Utility — Normalize Names

def normalize_name(name):
    name=name.strip()
    name=name.title()
    name=' '.join(name.split())
    return name

print(normalize_name(" alice johnson "))
print(normalize_name("BOB mcDONALD"))
```

```python
#Lab 7: Function Returning Multiple Values — Basic Stats

def basic_stats(numbers):
    total=sum(numbers)
    count=len(numbers)
    if count > 0:
       mean=total/count
    else:
        None
    return (count,total,mean)
    

count,total,mean=basic_stats([10,20,30 ])
print("Count:",count)
print("Total:",total)
print("Mean:",mean)
```

```python
#Lab 8: Apply Function to a List of Amounts — apply_tax

def apply_tax(amounts,tax_rate=0.1):
    new_list=[]
    for amount in amounts:
        #amounts=np.array(amounts)
        tax=amount*tax_rate
        tax_amount=amount+tax
        new_list.append(round(tax_amount,2))
    return new_list


print(apply_tax([100,200,350],0.05))
print("original list:",amounts)
```

```python
#Lab 9: Function with Optional Argument and Simple Validation

def process_score(score,max_score=100):
    
    if score < 0 or score > max_score:
        return None
    else:
        percent_score=(score/max_score)*100
        return percent_score

print(process_score(85))
print(process_score(150))

    
```

```python
#Lab 15: Final Integrated Lab — Sales ETL with Functions + NumPy

raw_sales = [120.0, 250.5, np.nan, -20.0, 400.0, 150.0, 0.0, 300.0]

def clean_sales(arr):
    np_arr=np.array(arr,dtype=float)
    np_arr[np_arr <= 0] = np.nan
    #return np_arr
    #print(np_arr)
    cleaned_array=np_arr[~np.isnan(np_arr)]
    return cleaned_array
    #print(cleaned_array)
    
            
clean_sales([120.0, 250.5, np.nan, -20.0, 400.0, 150.0, 0.0, 300.0])


def apply_discount_to_large_sales(arr,threshold=200,discount_rate=0.1):
        arr=np.array(arr,dtype=float)
        discounted=arr.copy()
        mask=arr > threshold
        #discount_rate=arr*0.1
        #discount_amount=arr-discount_rate
        discounted[mask] = arr[mask] * (1-discount_rate)
        cleaned=discounted[~np.isnan(discounted)]
        return cleaned
        #if arr > threshold:
            #discount_rate=arr*0.1
           # discount_amount=arr-discount_rate
            #return discount_amount
        #else:
           # discount_amount=arr
       # print("discount" , discount_amount)
print(apply_discount_to_large_sales(clean_sales([120.0, 250.5, np.nan, -20.0, 400.0, 150.0, 0.0, 300.0])))


def summarize_sales(arr):
    total=arr.sum()
    count=len(arr)
    Mean=arr.mean()
    Min=arr.min()
    Max=arr.max()
    return {
        "total":total,
        "count":count,
        "mean":Mean,
        "min":Min,
        "max":Max
    }

    

cleaned_array=clean_sales(raw_sales)
discounted_array=apply_discount_to_large_sales(cleaned_array,threshold=200,discount_rate=0.1)
summary=summarize_sales(discounted_array)

print(f"Total:{summary['total']:.2f}")
print(f"Count:{summary['count']:.2f}")
print(f"Mean:{summary['mean']:.2f}")
print(f"Min:{summary['min']:.2f}")
print(f"Max:{summary['max']:.2f}")

    
```
